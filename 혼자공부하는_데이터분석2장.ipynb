```python
#파이썬객체를 JSON 문자열로 변환하기
import json
```

```python
d_str = json.dumps(d, ensure_ascii=False) #json.dumps()함수는 아스키문자 외의 다른 문자를 16진수로 출력하기에 한글이 잘안보임
print(d_str)
```

```python
print(type(d_str))
```

```python
#JSON 문자열을 파이썬 객체로 변환하기
d2 = json.loads(d_str)
print(d2['name'])
```

```python
print(type(d2))
```

```python
x_str = """
<book>
  <name>혼자 공부하는 데이터 분석</name>
  <author>박해선</author>
  <year>2022</year>
</book>
"""
```

```python
import xml.etree.ElementTree as et
book = et.fromstring(x_str)
```

```python
print(type(book)) #fromstring()함수가 반환하는 객체는 단순한 파이썬 객체가 아니라 ElementTree 모듈 아래에 정의된 Element 클래스의 객체
```

```python
print(book.tag)
```

```python
book_childs = list(book)
print(book_childs)
```

```python
name, author, year = book_childs
print(name.text)
print(author.text)
print(year.text)
```

```python
#순서대로 자식 엘리먼트를 찾는 것은 위험. findtext()메서드를 사용해 자식 엘리먼트를 탐색하여 자동으로 텍스트를 반환
name = book.findtext('name')
author = book.findtext('author')
year = book.findtext('year')
print(name)
print(author)
print(year)
```

```python
x2_str = """
<books>
  <book>
    <name>혼자 공부하는 데이터 분석</name>
    <author>박해선</author>
    <year>2022</year>
  </book>
  <book>
    <name>혼자 공부하는 머신러닝+딥러닝</name>
    <author>박해선</author>
    <year>2020</year>
  </book>
</books>
"""
```

```python
#fromstring()함수를 사용해 부모 엘리먼트를 확인하면
books = et.fromstring(x2_str)
print(books.tag)
```

```python
#동일한 이름을 가진 여러 개의 자식 엘리먼트를 찾을 때는 finall()메서드와 for문을 함께 사용하면 편리
for book in books.findall('book'):
  name = book.findtext('name')
  author = book.findtext('author')
  year = book.findtext('year')
  print(name)
  print(author)
  print(year)
  print()
```

```python
def get_page_cnt(isbn):
  #yes24 도서 검색 페이지 URL
  url = 'http://www.yes24.com/Product/Search?domain=BOOK&query={}'
  #URL에 ISBN을 넣어 HTML을 가져온다
  r= requests.get(url.format(isbn))
  soup = BeautifulSoup(r.text, 'html.parser') #html 파싱
  #검색 결과에서 해당 도서를 선택한다.
  prd_info = soup.find('a',attrs = {'class':'gd_name'})
  if prd_info == None:
    return ''
  #도서 상세 페이지를 가져온다
  url = 'http://www.yes24.com'+prd_info['href']
  r= requests.get(url)
  soup = BeautifulSoup(r.text, 'html.parser')
  #품목정보<div>를 선택한다.
  prd_detail = soup.find('div',attrs={'id':'infoset_specific'})
  #테이블에 있는 <tr> 태그를 가져온다
  prd_tr_list = prd_detail.find_all('tr')
  #쪽수가 들어있는 <th>를 찾아 <td>에 담긴 값을 반환
  for tr in prd_tr_list:
    if tr.find('th').get_text() == '쪽수, 무게, 크기':
      return tr.find('td').get_text().split()[0]
  return ''
```
